```python
data = []
for index, row in df.iterrows():
    date = row['Date']
    year = pd.to_datetime(date).year
    month = pd.to_datetime(date).month
    day = pd.to_datetime(date).day
    weekday = pd.to_datetime(date).strftime('%A')
    holiday = False
    season = 'Unknown'
    cur.execute("SELECT date_key FROM Date_Dimension_Tab WHERE date = %s", (date,))
    existing_row = cur.fetchone()
    if existing_row:
        date_key = existing_row[0]
    else:
        print(f"Inserting row for {date}")
        cur.execute("INSERT INTO Date_Dimension_Tab (date, year, month, day, weekday, holiday, season) VALUES (%s, %s, %s, %s, %s, %s, %s)", (date, year, month, day, weekday, holiday, season))
        db.commit()
        cur.execute("SELECT date_key FROM Date_Dimension_Tab WHERE date = %s", (date,))
        date_key = cur.fetchone()[0]

    open_price = row['Open']
    high_price = row['High']
    low_price = row['Low']
    close_price = row['Close']
    volume = row['Total Trade Quantity']
    data.append((date_key, open_price, high_price, low_price, close_price, volume))

# Insert data into Stock Price Fact table
if data:
    query = "INSERT INTO Stock_Price_Fact_Tab (date_key, open_price, high_price, low_price, close_price, volume) VALUES (%s, %s, %s, %s, %s, %s)"
    cur.executemany(query, data)
    db.commit()

# Close the database connection and cursor
```
